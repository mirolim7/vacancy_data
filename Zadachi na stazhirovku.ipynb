## Задача 1. Распределяем таблицу данных по каждому полю

Определяем количество каждого из вакансий

```python
vacancies = {}

for profs in ds.iterrows():
    
    nm = profs[1]['name_vacancy']
    if nm in vacancies:
        vacancies[nm] += 1
    else:
        vacancies[nm] = 1
```

```python
vacancies
```

Так же определяем должности:

```python
duties = {}

def isNaN(n):
    return n != n

for profs in ds.iterrows():
    
    nm = profs[1]['duties']
    if not isNaN(nm):
        if nm in duties:
            duties[nm] += 1
        else:
            duties[nm] = 1
```

```python
duties
```

Требования:

```python
requirements = {}

for profs in ds.iterrows():
    
    nm = profs[1]['requirements']
    if not isNaN(nm):
        if nm in requirements:
            requirements[nm] += 1
        else:
            requirements[nm] = 1
```

```python
requirements
```

А также сферы деятельности:

```python
conditions = {}

for profs in ds.iterrows():
    
    nm = profs[1]['conditions']
    if not isNaN(nm):
        if nm in conditions:
            conditions[nm] += 1
        else:
            conditions[nm] = 1
```

```python
conditions
```

 ## Задача 2. Определяем самые востребованные навыки и т.п.

Находим самые популярные сферы деятельности:

```python
topD = dict(sorted(duties.items(), key=itemgetter(1), reverse=True)[:10])
```

```python
topD
```

Также и требования:

```python
topR = dict(sorted(requirements.items(), key=itemgetter(1), reverse=True)[:7])
```

```python
topR
```

Плюс ещё самые популярные сферы деятельности:

```python
topC = dict(sorted(conditions.items(), key=itemgetter(1), reverse=True)[:10])
```

```python
topC
```

Вакансии по самым популярным требованиям(requirements):

```python
topVR = {}

for n in ds.iterrows():
    
    i = n[1]['requirements']
    for key, value in topR.items():
        if i == key: 
            vc = n[1]['name_vacancy']
            if vc in topVR:
                topVR[vc] += 1
            else:
                topVR[vc] = 1
```

```python
topVR
```

Вакансии по самым популярным должностям(duties):

```python
topVD = {}

for n in ds.iterrows():
    
    i = n[1]['duties']
    for key, value in topD.items():
        if i == key: 
            vc = n[1]['name_vacancy']
            if vc in topVD:
                topVD[vc] += 1
            else:
                topVD[vc] = 1
```

```python
topVD
```

Вакансии по топ сферам деятельности(conditions):

```python
topVC = {}

for n in ds.iterrows():
    
    i = n[1]['conditions']
    for key, value in topC.items():
        if i == key: 
            vc = n[1]['name_vacancy']
            if vc in topVC:
                topVC[vc] += 1
            else:
                topVC[vc] = 1
```

```python
topVC
```

Вакансии по всем критериям(самые востребованные):

```python
for k, v in topVR.items():
    for key, value in topVD.items():
        for i, j in topVC.items():
            if k == key and k == i: 
                print(k)
```
